```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
HRattrition = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
HRattrition.info()
```

```python
HRattrition1 = HRattrition[['Attrition', 'JobRole', 'DistanceFromHome', 'MonthlyIncome', 'Education' ]]
HRattrition1
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=HRattrition1,x="JobRole", y="DistanceFromHome")
plt.xlabel("Job Role")
plt.ylabel("Distance From Home")
plt.title("Distance from home by job role and attrition.")
plt.color='red'
plt.show
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(data=HRattrition1,x="JobRole", y="DistanceFromHome")
plt.xlabel("Job Role")
plt.ylabel("Distance From Home")
plt.title("Distance from home by job role and attrition.")
plt.show
```

```python
HRattrition_monthlyincome = HRattrition1[['Attrition', 'MonthlyIncome', 'Education']]
HRattrition2 =HRattrition_monthlyincome.groupby(['Attrition','Education']).mean().reset_index()
HRattrition.head()
```

```python
HRattrition2 =HRattrition_monthlyincome.groupby(['Attrition','Education']).mean().reset_index()
HRattrition2
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Education", y="MonthlyIncome" ,hue="Attrition", data=HRattrition2)
plt.xlabel("Education",color="red")
plt.ylabel("Average MonthlyIncome",color="red")
plt.title("Average monthly income by education and attrition")
plt.show
```
